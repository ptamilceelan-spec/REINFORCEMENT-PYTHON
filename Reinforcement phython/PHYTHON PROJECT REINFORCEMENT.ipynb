```python
class Student:
    def __init__(self, roll_number, name, age, place, grade, marks=None):
        self.roll_number = roll_number
        self.name = name
        self.age = age
        self.place = place
        self.grade = grade
        self.marks = marks if marks else {}
        self.total_classes = 0
        self.attended_classes = 0
        
    def calculate_cgpa(self):
        if not self.marks:
            return 0.0
        avg = sum(self.marks.values()) / len(self.marks)
        return round(avg / 9.5, 2)

    def get_attendance_percentage(self):
        if self.total_classes == 0:
            return 0
        return round((self.attended_classes / self.total_classes) * 100, 2)

    def display_student_details(self):
        marks_str = ", ".join(f"{k}:{v}" for k, v in self.marks.items())
        return (
            "--------------------------------------------\n"
            f"Roll Number : {self.roll_number}\n"
            f"Name        : {self.name}\n"
            f"Age         : {self.age}\n"
            f"Place       : {self.place}\n"
            f"Grade       : {self.grade}\n"
            f"Marks       : {marks_str}\n"
            f"CGPA        : {self.calculate_cgpa()}\n"
            f"Attendance  : {self.get_attendance_percentage()}% "
            f"({self.attended_classes}/{self.total_classes})\n"
            "--------------------------------------------"
        )


class StudentManager:
    def __init__(self):
        self.students = []

    def add_student(self, student):
        if any(s.roll_number == student.roll_number for s in self.students):
            print("Student already exists.")
        else:
            self.students.append(student)
            print("Student added successfully.")

    def view_students(self):
        if not self.students:
            print("No students found.")
        for s in self.students:
            print(s.display_student_details())

    def find_student(self, roll_number):
        for s in self.students:
            if s.roll_number == roll_number:
                return s
        return None

    def search_students(self, roll_number):
        s = self.find_student(roll_number)
        print(s.display_student_details() if s else "Student not found.")

    def update_student(self, roll_number, new_details):
        s = self.find_student(roll_number)
        if not s:
            print("Student not found.")
            return
        for k, v in new_details.items():
            setattr(s, k, v)
        print("Student updated successfully.")

    def delete_student(self, roll_number):
        self.students = [s for s in self.students if s.roll_number != roll_number]
        print("Student deleted successfully.")

    def mark_attendance(self):
        if not self.students:
            print("No students available.")
            return
        for s in self.students:
            status = input(f"{s.name} present? (p/a): ").lower()
            s.total_classes += 1
            if status == 'p':
                s.attended_classes += 1
        print("Attendance marked.")



class Teacher:
    def __init__(self, teacher_id, name, age, place, subject):
        self.teacher_id = teacher_id
        self.name = name
        self.age = age
        self.place = place
        self.subject = subject

    def display_teacher_details(self):
        return f"{self.teacher_id} | {self.name} | {self.age} | {self.place} | {self.subject}"


class TeacherManager:
    def __init__(self):
        self.teachers = []

    def add_teacher(self, t):
        self.teachers.append(t)
        print("Teacher added.")

    def view_teachers(self):
        if not self.teachers:
            print("No teachers.")
        for t in self.teachers:
            print(t.display_teacher_details())

    def find_teacher(self, tid):
        for t in self.teachers:
            if t.teacher_id == tid:
                return t
        return None

    def update_teacher(self, tid, data):
        t = self.find_teacher(tid)
        if not t:
            print("Teacher not found.")
            return
        for k, v in data.items():
            setattr(t, k, v)
        print("Teacher updated.")

    def delete_teacher(self, tid):
        self.teachers = [t for t in self.teachers if t.teacher_id != tid]
        print("Teacher deleted.")


# ---------------- MAIN ----------------
def main():
    students = StudentManager()
    teachers = TeacherManager()

    while True:
        print("\n1.Add Student  2.View Students  3.Search Student  4.Update Student")
        print("5.Delete Student  6.Mark Attendance  7.Add Teacher  8.View Teachers")
        print("9.Update Teacher 10.Delete Teacher 15.Exit")

        choice = input("Choice: ")

        if choice == '1':
            s = Student(
                input("Roll: "), input("Name: "), input("Age: "),
                input("Place: "), input("Grade: ")
            )
            students.add_student(s)

        elif choice == '2':
            students.view_students()

        elif choice == '3':
            students.search_students(input("Roll: "))

        elif choice == '4':
            roll = input("Roll: ")
            students.update_student(roll, {'name': input("New Name: ")})

        elif choice == '5':
            students.delete_student(input("Roll: "))

        elif choice == '6':
            students.mark_attendance()

        elif choice == '7':
            t = Teacher(
                input("ID: "), input("Name: "), input("Age: "),
                input("Place: "), input("Subject: ")
            )
            teachers.add_teacher(t)

        elif choice == '8':
            teachers.view_teachers()

        elif choice == '9':
            teachers.update_teacher(input("ID: "), {'name': input("New Name: ")})

        elif choice == '10':
            teachers.delete_teacher(input("ID: "))

        elif choice == '15':
            print("Exited.")
            break

        else:
            print("Invalid choice.")


main()
```
